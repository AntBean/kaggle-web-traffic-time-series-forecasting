# Test Pandas Datetime Index

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
days = 2000
rows = 1000
```

```python
ini_date = '2010-01-01'
```

## 1.1 Numpy string method

```python
dtindex1 = np.arange(np.datetime64(ini_date), 
                     np.datetime64(ini_date) 
                     + np.timedelta64(days, 'D')).astype('str')
dtindex1
```

```python
df1 = pd.DataFrame(np.array([np.arange(days) for x in range(rows)]), 
                   columns=dtindex1)
```

```python
df1
```

```python
df1.columns
```

```python
df1['id'] = np.arange(rows)
```

```python
df1
```

```python
df1.info()
```

```python
df1_f = pd.melt(df1, id_vars='id', var_name='date', value_name='visits')
```

```python
df1_f
```

## 1.2 Pandas date_range method

```python
dtindex2 = pd.date_range(ini_date,
                         periods=np.timedelta64(days,'D'), 
                         freq = 'D', unit = 'D')
dtindex2
```

```python
df2 = pd.DataFrame(np.array([np.arange(days) for x in range(rows)]), 
                   columns=dtindex2)
```

```python
df2
```

```python
df2.columns
```

```python
df2['id'] = np.arange(rows)
```

```python
df2
```

```python
df2.info()
```

```python
df2_f = pd.melt(df2, id_vars='id', var_name='date', value_name='visits')
```

```python
df2_f
```
