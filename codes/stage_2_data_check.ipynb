```python
output_dates_all = output_df.date.values.astype('datetime64[D]')
```

```python
output_first_day = str(output_dates_all.min())
output_final_day = str(output_dates_all.max())
output_first_day, output_final_day
```

```python
from datetime import datetime
def weekday(datestr):
    return datetime.strptime(datestr,'%Y-%m-%d').weekday()
def daydiff(dstr1,dstr2):
    return datetime.strptime(dstr1,'%Y-%m-%d') - datetime.strptime(dstr2,'%Y-%m-%d') 
```

```python
daydiff(output_first_day, output_final_day)
```
